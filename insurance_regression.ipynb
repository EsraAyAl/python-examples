```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = insurance_data.iloc[:,:1].values
y = insurance_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b') #colour is blue
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
X = insurance_data['age'].values.reshape(-1, 1) # create a 2D array
y = insurance_data['charges'].values
```

```python
# Fit and predict
charges_model = LinearRegression()
charges_model.fit(X, y)
y_pred = charges_model.predict(X)
```

```python
# Plot model and data
plt.figure()
sns.scatterplot(data=insurance_data, x='age', y='charges')
plt.plot(X, y_pred)
plt.show()
plt.close()
```

```python
# Predict insurance charge according to assigned age

unk_x = [[38]] # Assigned age as a variable

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

print("Charge should be:", charges_model.predict(unk_x))
```
